```python
import pandas as pd

df = pd.read_csv("predictions.csv")
print(df.head())
print(df.columns)
```

```python
print("XGBoost predicted frauds:", df['xgb_pred'].sum())
print("IsolationForest anomalies:", df['isolation_anomaly'].sum())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
df['xgb_pred'].value_counts().plot(kind='bar', title='XGBoost Predicted Fraud')
plt.show()

plt.figure(figsize=(6,4))
df['isolation_anomaly'].value_counts().plot(kind='bar', title='IsolationForest Anomalies')
plt.show()
```
